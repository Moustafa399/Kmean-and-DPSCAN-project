```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:\subjects cs 4\course AI\datasets\dataset6.csv',header=None,names=['X1','X2'])
#df=df.rename(columns={df.columns[0]:'x',df.rename(columns={df.columns[0]:'y'})
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='X1', y='X2')

# Display the plot
plt.show()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# K-means clustering
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(df)

# DBSCAN clustering
dbscan = DBSCAN(eps=0.1, min_samples=3)
dbscan_labels = dbscan.fit_predict(df)

# Plot the data and cluster labels
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].scatter(df.iloc[:, 0], df.iloc[:, 1], c=kmeans_labels)
axs[0].set_title('K-means Clustering')

axs[1].scatter(df.iloc[:, 0], df.iloc[:, 1], c=dbscan_labels)
axs[1].set_title('DBSCAN Clustering')

plt.show()
```

```python
df2 = pd.read_csv(r'E:\subjects cs 4\course AI\datasets\dataset7.csv',header=None,names=['n1','n2'])
df2.head()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# K-means 
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(df2)

# DBSCAN 
dbscan = DBSCAN(eps=0.2, min_samples=2)
dbscan_labels = dbscan.fit_predict(df2)


fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].scatter(df.iloc[:, 0], df2.iloc[:, 1], c=kmeans_labels)
axs[0].set_title('K-means Clustering')

axs[1].scatter(df.iloc[:, 0], df2.iloc[:, 1], c=dbscan_labels)
axs[1].set_title('DBSCAN Clustering')

plt.show()
```
